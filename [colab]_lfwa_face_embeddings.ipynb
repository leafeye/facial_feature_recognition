```
import pandas as pd
import numpy as np
import os
from tqdm import tqdm
```

```
lfwa_labels = pd.read_csv('/content/drive/My Drive/Colab/lfwa/lfwa_labels.csv')
```

```
lfwa_labels.head()
```

```
files = lfwa_labels.file
```

```
files[:5]
```

```
img_folder = '/content/drive/My Drive/Colab/lfwa/lfw/'
```

```
img_folder + files[0]
```

```
lfwa_embed_list = []
lfwa_file_list = []

for file in tqdm(files):
  img_path = img_folder + file
  
  #load image
  img = face_recognition.load_image_file(img_path)
  
  #locate face in image, use the cnn method
  face_loc = face_recognition.face_locations(img, model='cnn')
  
  #if no face or more than one face is detected, skip
  if len(face_loc) != 1:
    print("error for: " + file + " number of faces detected: " + str(len(face_loc)) )
    
    face_loc = face_recognition.face_locations(img, model='hog')
    
    if len(face_loc) != 1:
      print('still an error, skip')
      continue
    
  #encode face to 128d embedding
  embedding = face_recognition.face_encodings(img, known_face_locations=face_loc)[0]
  
  #append to list
  lfwa_embed_list.append(embedding)
  lfwa_file_list.append(file)
```

```
len(lfwa_file_list)
```

```
len(lfwa_embed_list)
```

```
#also store in pandas dataframe, so the corresponding filenames can't get mixed up
lfwa_embeddings_pandas = pd.DataFrame({'file': lfwa_file_list, 'encodings': lfwa_embed_list })
```
