```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline  

import math
```

```python
def face_rating_data_expand(fr_df):
    """"this function creates a column with correct name in the pandas DF 
    for every key in the demographics and responses json section """
    
    for key in fr_df.demographics.iloc[0].keys():
        fr_df['participant_'+ key] = fr_df.apply(lambda row: row.demographics[key], axis=1) 

    for key in fr_df.responses.iloc[0].keys():
        fr_df['face_perceived_'+ key] = fr_df.apply(lambda row: row.responses[key], axis=1) 
```

```python
#load face rating data 1 json file in pandas
fr_1 = pd.read_json('parsed_json/face-rating-data-1.json')
```

```python
fr_1.head(2)
```

```python
#expand the dataframe dictionaries to new columns
face_rating_data_expand(fr_1)
```

```python
fr_2 = pd.read_json('parsed_json/face-rating-data-2.json')
```

```python
fr_2.head(2)
```

The second face rating datafile has empty values for the ethnicity column.

We need a change those to np.nan values before we can run the face_rating_data_expand function

```python
def make_empty_ethn_nan(df):
    for index, row in df.iterrows():
        if 'ethnicity' not in row.responses.keys():
            row.responses['ethnicity'] = np.nan
    
```

```python
#change empty values for ethnicity to np.nan
make_empty_ethn_nan(fr_2)

#expand the dataframe dictionaries to new columns
face_rating_data_expand(fr_2)
```
